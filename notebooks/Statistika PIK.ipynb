## Турклуб ПИК

```python
with open('../data/pik_parsed_routes_20210529_2220.txt','r',encoding='utf-8') as f:
    pik_routes = f.readlines()
```

```python
pik_routes_evaled = [x for x in map(lambda x: eval(x), pik_routes)]
```

```python
pik_df = pd.DataFrame(columns = ['url','name','region','distance','duration','level','type','is_children',
                       'short_desc_len', 'full_desc_len','num_images','num_guides','num_hikes'
                       ])
```

```python
for route in pik_routes_evaled:
    pik_df.loc[pik_df.shape[0],:] = [
        route['url'], route['name'], route['region'],
        route['distance'], route['duration'], route['level'],
        route['type'], route['is_children'], len(route['short_description']),
        len(route['full_description']), len(route['images']), len(route['guides']), len(route['hikes'])
                                  ]
```
